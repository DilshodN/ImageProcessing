```python
import numpy as np
import cv2
from IPython.display import display
from PIL import Image
import matplotlib.pyplot as plt
import random
```

```python
def show(img):
    plt.imshow(img, cmap='gray')
```

```python
img = cv2.imread('/Users/dilshod/Downloads/lena.png')

img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

show(img)
```

```python
def foo(a, minv, maxv):
    if a > maxv:
        return maxv
    if a < minv:
        return minv
    return a
```

```python
def addNoise(origImage: np.array, s=150):
    image = origImage.copy()
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            noise = np.random.uniform(-s, s)
            new_val = image[i, j] + noise
            image[i, j] = foo(new_val, 0, 255)
    return image
```

```python
noisy = addNoise(img)
show(noisy)
```

```python
def makeNoisyImage(origImage: np.array, N: int):
    noisyImages = []
    for i in range(N):
        noisyImages.append(addNoise(origImage))
    return np.array(noisyImages)
```

```python
LennaImages = makeNoisyImage(img, 10)
```

```python
filtered = np.mean(LennaImages, axis=0)
```
